```python
def check_game_over_3(markers):
  game_over = False
  tie = False
  winner =0
  x_pos = 0
  for x in markers:
    #check columns
    if sum(x) == 3:
      winner = 1
      game_over = True
    if sum(x) == -3:
      winner = 2
      game_over = True
    
    #check rows
    if markers[0][x_pos] + markers [1][x_pos] + markers [2][x_pos] == 3:
      winner = 1
      game_over = True
    if markers[0][x_pos] + markers [1][x_pos] + markers [2][x_pos] == -3:
      winner = 2
      game_over = True
    x_pos += 1

  #check cross
  if markers[0][0] + markers[1][1] + markers [2][2] == 3 or markers[2][0] + markers[1][1] + markers [0][2] == 3:
    winner = 1
    game_over = True
  if markers[0][0] + markers[1][1] + markers [2][2] == -3 or markers[2][0] + markers[1][1] + markers [0][2] == -3:
    winner = 2
    game_over = True

  #check for tie
  if game_over == False:
    tie = True
    for row in markers:
      for i in row:
        if i == 0:
          tie = False
    #if it is a tie, then call game over and set winner to 0 (no one)
    if tie == True:
      game_over = True
      winner = 0
  return game_over,winner,tie
  # if game_over == False:
```

```python
beta = 0.2
gamma = 0.1
s_i = STATES[0].copy()
wd = 0
while True:
  Q_temp = Q.copy()
  a = np.where(s_i==0)[0]
  if len(a) == 0:
    a = range(9)
  a = random.choice(a)
  ind = np.where((STATES==s_i).all(-1))[0][0] # index of current state
  ind2 = np.random.choice(np.arange(STATES_SPACE_SIZE),1,p= transition_probability[ind][a])[0]
  R = REWARDS[ind2]
  maxb = np.max(Q[ind2])
  Q[ind][a] += beta*(R + gamma*(maxb) - Q[ind,a])
  s_i = STATES[ind2].copy()
  print(np.max(np.absolute(Q_temp - Q) ) )
  if np.max(np.absolute(Q_temp - Q))<0.0001:
    wd+=1
  else:
    wd = 0
  if wd == 10:
    break
```

```python
pi_star = [0]*len(STATES)
for i in range(3381):
  
  pi_star[i] = np.argmax(Q[i])
```
